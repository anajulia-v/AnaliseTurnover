## 🧾 **Conclusão**

A análise exploratória dos dados de Recursos Humanos permitiu identificar fatores
relevantes associados à rotatividade de funcionários, evidenciando que o desligamento
não ocorre de forma aleatória, mas está relacionado a características organizacionais,
comportamentais e de satisfação.

Os resultados indicam que aspectos ligados à progressão de carreira, como longos períodos
sem promoção, estão associados a maiores taxas de desligamento. Da mesma forma, fatores
remuneratórios, incluindo baixos percentuais de aumento salarial e faixas salariais mais
baixas, apresentam relação consistente com maior rotatividade.

Em relação à qualidade de vida e carga de trabalho, observou-se que a realização frequente
de horas extras e a alta frequência de viagens estão fortemente associadas a maiores índices
de saída, sugerindo impactos negativos de sobrecarga e deslocamentos frequentes na retenção
dos funcionários. Por outro lado, a distância até o trabalho, analisada isoladamente, não
se mostrou um fator determinante.

A satisfação dos funcionários (e-sat) destacou-se como uma das variáveis mais relevantes,
com taxas de desligamento significativamente mais elevadas entre colaboradores com baixos
níveis de satisfação, reforçando a importância de ações voltadas à experiência do funcionário.

De maneira geral, os achados reforçam que estratégias de retenção devem priorizar políticas
de desenvolvimento de carreira, reconhecimento, equilíbrio entre vida pessoal e trabalho
e melhoria da satisfação dos colaboradores, contribuindo para a redução do turnover.
